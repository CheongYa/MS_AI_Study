```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
```

```python
# 기존 데이터 닥스훈트와 진돗개
dach_length = [55, 57, 64, 63, 58, 49, 54, 61]
dach_height = [30, 31, 36, 30, 33, 25, 37, 34]
 
jin_length = [56, 47, 56, 46, 49, 53, 52, 48]
jin_height = [52, 52, 50, 53, 50, 53, 49, 54]

# 새로운 데이터
newdata_lenth = [59]
newdata_height = [35]
```

```python
plt.scatter(dach_length, dach_height, label='Dachshund') # 파랑색
plt.scatter(jin_length, jin_height, label='JindoDog') # 주황색
plt.scatter(newdata_lenth, newdata_height, marker='p') # 녹색
plt.legend(loc='upper right')
plt.show()
```

```python
# 데이터 테이블로 만들기
d_data = np.column_stack((dach_length, dach_height))
d_label = np.zeros(len(d_data)) # d_data의 길이 만큼 0으로 채워진 행렬 생성

j_data = np.column_stack((jin_length, jin_height))
j_label = np.ones(len(j_data))

# 두 데이터 합치기(데이터 결합)
dogs = np.concat((d_data, j_data))
labels = np.concat((d_label, j_label))

dogs, labels
```

```python
k = 3
knn = KNeighborsClassifier(n_neighbors=k)
knn
```

```python
knn.fit(dogs, labels)
```

```python
# 새로운 데이터 합치기
new_data = np.column_stack((newdata_lenth, newdata_height))
```

```python
# 예측값
y_pred = knn.predict(new_data)
y_pred
```

```python
dog_classes = {0 : '닥스훈트', 1 : '진돗개'}

print(f'예측데이터 : {new_data}, 예측결과 : {dog_classes[y_pred[0]]}')
```

```python
# 새로운 데이터2
newdata_lenth_2 = [50]
newdata_height_2 = [41]
```

```python
new_data_2 = np.column_stack((newdata_lenth_2, newdata_height_2))
# new_data_2 = np.array([50, 41]) 으로도 가능
```

```python
for k in [1, 3, 5, 9]:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(dogs, labels)
    y_pred2 = knn.predict(new_data_2)
    print(f'k={k}일 때 예측데이터 : {new_data_2}, 예측결과 : {dog_classes[y_pred2[0]]}')
```
