```python
# matplotlib.pyplot를 사용하여 시각화 그래프 생성

import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel("data/fine_dust.xlsx", index_col='area')
data
```

```python
data2018 = data[2018]
plt.figure(figsize=(15, 4)) # 크기
plt.plot(data2018, color='b', marker='o') # 그래프
plt.xlabel('area') # x축 가로열
plt.ylabel('micrometer') # y축 세로열
plt.title('2018 fine dust line graph') # 상단 타이틀
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(15, 4))

for year in range(2016, 2020):
    chartdata = data[year]
    plt.plot(chartdata, marker='o', label=str(year)) # 그래프

plt.xlabel('area') # x축 가로열
plt.ylabel('micrometer') # y축 세로열
plt.title('2016~2019 fine dust line graph') # 상단 타이틀
plt.legend()
plt.grid()
plt.show()
```

```python
# 막대그래프

data2017 = data[2017]
plt.figure(figsize=(15, 4)) # 크기
plt.bar(data2017.index, data2017, color='g') # 그래프, bar 사용, (x축, y축, 색)
plt.ylim(35, 55) # y축 시작과 끝 값
plt.xlabel('area') # x축 가로열
plt.ylabel('micrometer') # y축 세로열
plt.title('2017 fine dust line graph') # 상단 타이틀
plt.grid()
plt.show()
```

```python
# 산점도 그래프
data2 = pd.read_excel("data/health_screenings_2020_1000ea.xlsx")
```

```python
data2.head()
```

```python
heightdata = data2['height']
weightdata = data2['weight']

plt.figure(figsize=(15, 6))

plt.scatter(heightdata, weightdata)
```

```python
# 수축기 혈압, 이완기 혈압, 공복 혈당 그룹 산점도 그래

systolic_data = data2['systolic']
diastolic_data = data2['diastolic']
blood_sugar_data = data2['blood_sugar']

plt.figure(figsize=(10, 6))

plt.scatter(systolic_data, diastolic_data, color = 'r', label = 'systolic * diastolic')
plt.scatter(systolic_data, blood_sugar_data, color = 'g', label = 'systolic * blood_sugar_data')
plt.scatter(diastolic_data, blood_sugar_data, color = 'b', label = 'diastolic_data * blood_sugar_data')

plt.xlim(40, 180)
plt.ylim(30, 300)
plt.title('2020 Health Screening Group Scatter Graph')
plt.legend()
plt.show()
```

```python
# 여성의 성별, 키 데이터로 여성 키 히스토그램 생성

# 데이터셋 생성
female_data = data2.loc[data2['gender'] == 2][['gender', 'height']]

plt.figure(figsize=(10, 6))
# bins = 보여지는 개체의 수
plt.hist(female_data['height'], bins = 7, label = 'Female')

plt.xlim(130, 180)
plt.title('2020 Health Screening Femail Data Histogram Graph')
plt.legend()
plt.grid()
plt.show()
```

```python
# 그룹 히스토그램 생성
# 여성, 남성 키 히스토그램

# 데이터셋 생성
male_data = data2.loc[data2['gender'] == 1][['gender', 'height']]
female_data = data2.loc[data2['gender'] == 2][['gender', 'height']]

plt.figure(figsize=(10, 6))
# alpha = 투명도
plt.hist(male_data['height'], alpha = 0.7, bins = 7, label = 'male')
plt.hist(female_data['height'], alpha = 0.5, bins = 7, label = 'Female')

plt.xlim(130, 200)
plt.title('2020 Health Screening Femail Data Histogram Graph')
plt.legend()
plt.grid()
plt.show()
```

```python
# 상자수염 그래프 작성

import numpy as np

height = np.array([male_data['height'], female_data['height']], dtype=object)
plt.figure(figsize=(10, 6))
plt.boxplot(height, tick_labels=['male', 'female'])

plt.ylim(130, 200)
plt.title('2020 Health Screening Femail Data Histogram Graph')
plt.show()
```
