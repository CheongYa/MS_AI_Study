```python
# html 링크를 토대로 읽은 후 줄 수 출력

import matplotlib.pyplot as plt
import pandas as pd

table = "https://ko.wikipedia.org/wiki/%EB%8C%80%ED%95%9C%EB%AF%BC%EA%B5%AD%EC%9D%98_%EC%9D%B8%EA%B5%AC"

tdata = pd.read_html(table)

len(tdata)
```

```python
df1 = tdata[4] # 광복 이전 데이터
df2 = tdata[5] # 광복 이후 데이터
```

```python
# 광복 이전과 광복 이후 데이터 병합, 행 방향으로 병합
df = pd.concat([df1, df2])
df
```

```python
# 비어있는 곳(True, NaN)을 값(value, 0)으로 채운다.
df.fillna(value=0, inplace=True)

meanData = df.groupby('연도 (년)')[['합계출산율']].mean()
plt.plot(meanData)
plt.show()
```
